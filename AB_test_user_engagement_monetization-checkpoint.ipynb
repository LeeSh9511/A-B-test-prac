```python
%matplotlib inline
import pandas as pd
import numpy as np
from scipy.stats import norm
from math import sqrt
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm,mannwhitneyu,levene
```

```python
test_data=pd.read_csv('c:/Users/USER/Desktop/abtest/ab2_repo/Test_Group_Data.csv')
control_data=pd.read_csv('c:/Users/USER/Desktop/abtest/ab2_repo/Control_Group_Data.csv')
```

```python
# gender에 따른 차이는 크지 않아 보임.
print(control_data.groupby('Gender')['Average_Purchase_Value'].mean())
print(test_data.groupby('Gender')['Average_Purchase_Value'].mean())
```

```python
# Calculate Engagement and Monetization Scores for control
control_data['Engagement_Score'] = (control_data['Sessions'] + control_data['Time_Spent_Minutes'] + 
                                    control_data['Actions_Per_Session']) / 3
control_data['Monetization_Score'] = (control_data['Purchases'] * control_data['Average_Purchase_Value']) / 10
# Calculate Engagement and Monetization Scores for test
test_data['Engagement_Score'] = (test_data['Sessions'] + test_data['Time_Spent_Minutes'] + 
                                 test_data['Actions_Per_Session']) / 3
test_data['Monetization_Score'] = (test_data['Purchases'] * test_data['Average_Purchase_Value']) / 10
# Combine into final dataset
ab_test_data = pd.concat([control_data, test_data], ignore_index=True)
```

- 유저 참여 정도와 과금 정도를 의미하는 score를 정의함.
-  monetization score를 정의할 때와 같이, 임의로 값을 나눔으로써 데이터의 scale을 조정할 땐 주의가 필요함.

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=ab_test_data, x='Engagement_Score', hue='Group', kde=True, bins=50, palette='muted')
plt.title('Distribution of Engagement Score by Group')
plt.xlabel('Engagement Score')
plt.ylabel('Frequency')
plt.show()
```

```python
levene_stat,p_value=levene(ab_test_data[ab_test_data['Group'] == 'Control']['Engagement_Score'],ab_test_data[ab_test_data['Group'] == 'Test']['Engagement_Score'])
if p_value < 0.05:
    print(" Control group과 Test group의 engagement score는 등분산 가정이 불가하다.")
else:
    print(" Control group과 Test group의 engagement score는 등분산 가정이 가능하다. ")
```

- test group에서 engagement score가 높게 나타남.
- 분포가 normal distribution을 만족하는 것으로 보임. t-test로 control group과 test group간 차이를 검정.
- 등분산 가정을 만족하지만, 분산 pooling 과정에서의 왜곡 가능성을 줄이기 위해 Welch's t-test(이분산 t-test)를 사용.

```python
# Plot distribution of Monetization Score for both groups
plt.figure(figsize=(8, 6))
sns.histplot(data=ab_test_data, x='Monetization_Score', hue='Group', kde=True, bins=50, palette='muted')
plt.title('Distribution of Monetization Score by Group')
plt.xlabel('Monetization Score')
plt.ylabel('Frequency')
plt.show()
```

```python
observed_diff=test_data['Monetization_Score'].mean()-control_data['Monetization_Score'].mean()
boot_diffs = []
for _ in range(10000):
    test_ms = np.random.choice(test_data['Monetization_Score'].tolist(), size=len(test_data), replace=True)
    control_ms = np.random.choice(control_data['Monetization_Score'].tolist(), size=len(control_data), replace=True)
    boot_diffs.append(test_ms.mean() - control_ms.mean())
boot_diffs = np.array(boot_diffs)
```

```python
# 95% 신뢰구간
ci_lower = np.percentile(boot_diffs, 2.5)
ci_upper = np.percentile(boot_diffs, 97.5)

print(f"95% Confidence Interval: [{ci_lower:.4f}, {ci_upper:.4f}]")
```

```python
plt.hist(boot_diffs, bins=50, edgecolor='black')
plt.axvline(ci_lower, color='red', linestyle='--', label='95% CI Lower')
plt.axvline(ci_upper, color='red', linestyle='--', label='95% CI Upper')
plt.axvline(observed_diff, color='blue', linestyle='-', label='Observed Diff')
plt.title('Bootstrap Distribution of Monetization score Difference')
plt.xlabel('Difference in Monetization score (test - control)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

- test group에서 monetization score이 높게 나타남.
- right-skewed distribution을 보임. 정규성 가정을 만족하지 못하므로, 
- 비모수 검정법인 Wilcoxon Rank-SUM test로 cotrol group과 test group간 차이를 검정
- Test 그룹은 Control 그룹에 비해 평균적으로 약 0.33 만큼 높은 Monetization Score를 보였으며, test 그룹과 control 그룹의 차이에 대한 95% bootstrap confidence interval은 [0.275, 0.375]. => test group의 monetization score가 적어도 0.275, 많게는 0.375 높다고 주장할 수 있다.

### statistical hypothesis test
- Welch’s t-test(이분산 t-test)
- Wilcoxon Rank-Sum Test (Mann-Whitney U Test)

```python
# Recalculating means and standard deviations using the dataset from EDA
mean_control_engagement = ab_test_data[ab_test_data['Group'] == 'Control']['Engagement_Score'].mean()
mean_test_engagement = ab_test_data[ab_test_data['Group'] == 'Test']['Engagement_Score'].mean()
std_control_eng=np.sqrt(np.mean((ab_test_data[ab_test_data['Group'] == 'Control']['Engagement_Score']-mean_control_engagement)**2))
std_test_eng=np.sqrt(np.mean((ab_test_data[ab_test_data['Group'] == 'Test']['Engagement_Score']-mean_test_engagement)**2))

## monetization
u_stat,p_value_monetization=mannwhitneyu(ab_test_data[ab_test_data['Group'] == 'Test']['Monetization_Score'],ab_test_data[ab_test_data['Group'] == 'Control']['Monetization_Score'],alternative='greater')
##
n1=10000 # sample size of control dataset
n2=10000 # sample size of test dataset
## z-statistic for engagement score
z_score_engagement=(mean_test_engagement-mean_control_engagement)/np.sqrt(std_control_eng**2/n1+std_test_eng**2/n2)
p_value_engagement = 1 - norm.cdf(z_score_engagement)  # One-tailed test
print(z_score_engagement,p_value_engagement)
print(p_value_monetization)

```

```python
# Summarizing results
results = {
    "Metric": ["Monetization Score", "Engagement Score"],
    'Test' : ['Wilcoxon Rank-Sum Test','Welchs t-test'],
    "Test statistic": [u_stat, z_score_engagement],
    "P-Value": [p_value_monetization, p_value_engagement],
    "Significant (α=0.05)": [p_value_monetization < 0.05, p_value_engagement < 0.05],
     "Conclusion": [
        "Reject Null Hypothesis" if p_value_monetization < 0.05 else "Fail to Reject Null Hypothesis",
        "Reject Null Hypothesis" if p_value_engagement < 0.05 else "Fail to Reject Null Hypothesis"
    ]
}
# Displaying the results to the user
results_df = pd.DataFrame(results)

results_df
```

- 이는 새로운 시도가 유저들의 게임 몰입도를 높이고, 결제 활동의 유의미한 증가로 이어진다는 것을 의미한다고 볼 수 있음.

- $t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$
